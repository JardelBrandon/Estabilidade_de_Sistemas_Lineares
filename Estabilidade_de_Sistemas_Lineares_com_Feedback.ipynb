# Estabilidade de Sistemas Lineares com Realimentação

A estabilidade dos sistemas de feedback de malha fechada é fundamental para o projeto de sistema de controle. A estabilidade de um sistema deve exibir uma saída limitada se a entrada correspondente for limitada. Isso é conhecido como estabilidade de entrada limitada de saída. A estabilidade de um sistema com feedback está diretamente relacionada a localização das raízes da equação característica da função de transferência do sistema e para a localização dos valores próprios da matriz do sistema para um sistema em estado variado. O método Routh-Hurwitz é apresentado como uma ferramenta útil para avaliar estabilidade do sistema. A técnica nos permite calcular o número de raízes da equação característica no meio plano direito sem realmente computar os valores das raízes. Isso nos dá um método de design para determinar os valores de determinado sistema com parâmetros que levarão à estabilidade do circuito fechado.

## 1. O Conceito de Estabilidade

Ao considerar o projeto e análise de sistemas de controle de feedback, a estabilidade é de a maior importância. Do ponto de vista prático, um sistema de feedback de circuito fechado que é instável tem pouco valor. Tal como acontece com todas as declarações gerais, há exceções; mas para nossos propósitos, declararemos que todos os nossos projetos de controle devem resultar em um sistema estável de circuito fechado. Muitos sistemas físicos são inerentemente unsta- de loop aberto e alguns sistemas são até mesmo projetados para serem instáveis ​​em malha aberta. Mais moderno aeronaves de caça são de circuito aberto instáveis ​​por projeto e sem controle de feedback ativo auxiliando o piloto, eles não podem voar. O controle ativo é introduzido por engenheiros para estabilizar lize o sistema instável - isto é, a aeronave - para que outras considerações, como desempenho transitório, pode ser resolvido. Usando feedback, podemos estabilizar a instabilidade sistemas e, em seguida, com uma seleção criteriosa de parâmetros do controlador, podemos ajustar o desempenho transitório. Para sistemas estáveis ​​de malha aberta, ainda usamos feedback para ajuste o desempenho do circuito fechado para atender às especificações do projeto. Estes especifi-cátions tomam a forma de erros de rastreamento de estado estacionário, porcentagem de ultrapassagem, tempo de acomodação, tempo para atingir o pico e os outros índices discutidos nos Capítulos 4 e 5.

Podemos dizer que um sistema de realimentação em malha fechada é estável ou instável. Este tipo de caracterização estável / não estável é referido como estabilidade absoluta. O sistema possuindo estabilidade absoluta é chamado de sistema estável - o rótulo de absoluto é desistiu. Dado que um sistema de malha fechada é estável, podemos caracterizar ainda mais o grau de estabilidade. Isso é conhecido como estabilidade relativa. Os pioneiros do design de aeronaves estavam familiarizados com a noção de estabilidade relativa - quanto mais estável uma aeronave, o mais difícil era manobrar (isto é, virar). Um resultado da instituição relativabilidade dos caças modernos é alta manobrabilidade. Uma aeronave de caça é menos estável do que um transporte comercial, portanto, pode manobrar mais rapidamente. Na verdade, os movimentos de uma aeronave de caça pode ser bastante violento para os "passageiros". Como discutiremos mais tarde em nesta seção, podemos determinar que um sistema é estável (no sentido absoluto) por deter- garimpando que todos os pólos da função de transferência estão na metade esquerda do plano s, ou equivalentemente, que todos os autovalores da matriz do sistema A encontram-se na metade esquerda do plano s. Dado que todo o pólos (ou valores próprios) estão na metade esquerda .v plano, investigamos a estabilidade relativa por examinando as localizações relativas dos pólos (ou autovalores).

Um **sistema estável** é definido como um sistema com uma resposta de sistema limitada (limitada). Ou seja, se o sistema está sujeito a uma entrada limitada ou perturbação e a resposta é limitado em magnitude, o sistema é considerado estável.

**<center>Um sistema estável é um sistema dinâmico com uma resposta limitada a uma entrada limitada.</center>**

## 2. O Critério de Estabilidade de Routh-Hurwitz

A discussão e determinação da estabilidade tem ocupado o interesse de muitos engenheiros. Maxwell e Vyshnegradskii consideraram pela primeira vez a questão da estabilidade de sistemas dinâmicos. No final de 1800, A. Hurwitz e E. J. Routh independentemente publicou um método de investigação da estabilidade de um sistema linear [6, 7]. O O método de estabilidade de Routh-Hurwitz fornece uma resposta à questão da estabilidade por considerando a equação característica do sistema. A equação característica em a variável Laplace é escrita como

\begin{equation}
    \Delta(s)=q(s)=a_{n} s^{n}+a_{n-1} s^{n-1}+\cdots+a_{1} s+a_{0}=0 
\end{equation}

Para verificar a estabilidade do sistema, é necessário determinar se algum das raízes de q (s) encontra-se na metade direita do plano s. Se a Equação acima for escrita em forma fatorada, nós temos

\begin{equation}
    a_{n}\left(s-r_{1}\right)\left(s-r_{2}\right) \cdots\left(s-r_{n}\right)=0
\end{equation}

onde $r_i = $ enésima raiz da equação característica. Multiplicando os fatores juntos, nós encontramos isso

\begin{equation}
    \begin{aligned}
        q(s)=& a_{n} s^{n}-a_{n}\left(r_{1}+r_{2}+\cdots+r_{n}\right) s^{n-1} \\
        &+a_{n}\left(r_{1} r_{2}+r_{2} r_{3}+r_{1} r_{3}+\cdots\right) s^{n-2} \\
        &-a_{n}\left(r_{1} r_{2} r_{3}+r_{1} r_{2} r_{4} \cdots\right) s^{n-3}+\cdots \\
        &+a_{n}(-1)^{n} r_{1} r_{2} r_{3} \cdots r_{n}=0
    \end{aligned}
\end{equation}

Em outras palavras, para uma equação de enésimo grau, obtemos

\begin{equation}
    \begin{aligned}
        q(s)=& a_{n} s^{n}-a_{n}\left(\text{soma de todas as raízes}\right) s^{n-1} \\
        &+a_{n}\left(\text{soma dos produtos das raízes pegando 2 de cada vez}\right) s^{n-2} \\
        &-a_{n}\left(\text{soma dos produtos das raízes pegando 3 de cada vez}\right) s^{n-3}+\cdots \\
        &+ \cdots +a_{n}(-1)^{n} \left(\text{produtos de todas as n raízes}\right) = 0
    \end{aligned}
\end{equation}

Examinando as duas equações acima, notamos que todos os coeficientes do polinômio deve ter o mesmo sinal se todas as raízes estiverem no plano do lado esquerdo. Além disso, é necessário que todos os coeficientes para um sistema estável sejam diferentes de zero. Esses requisitos são necessário, mas não suficiente. Ou seja, sabemos imediatamente que o sistema é instável se eles não estão satisfeitos; no entanto, se eles estiverem satisfeitos, devemos prosseguir para averiguar a estabilidade do sistema. Por exemplo, quando a equação característica é

\begin{equation}
    q(s)=(s+2)\left(s^{2}-s+4\right)=\left(s^{3}+s^{2}+2 s+8\right)
\end{equation}

o sistema é instável, mas o polinômio possui todos os coeficientes positivos. O critério de Routh-Hurwitz é um critério necessário e suficiente para o estado bilidade dos sistemas lineares. O método foi originalmente desenvolvido em termos de determinação nants, mas usaremos a formulação de array mais conveniente. O critério de Routh-Hurwitz é baseado na ordenação dos coeficientes dos caracteres equação acterística

\begin{equation}
    a_{n} s^{n}+a_{n-1} s^{n-1}+a_{n-2} s^{n-2}+\cdots+a_{1} s+a_{0}=0
\end{equation}

em uma matriz ou programação da seguinte forma

\begin{equation}
    \begin{array}{l}
        s^{n} \\
        s^{n-1}
        \end{array} \mid \begin{array}{lll}
        a_{n} & a_{n-2} & a_{n-4} \cdots \\
        a_{n-1} & a_{n-3} & a_{n-5} \cdots
    \end{array}
\end{equation}

Outras linhas da programação são concluídas conforme

\begin{equation}
    \begin{array}{l|lll}
        s^{n} & a_{n} & a_{n-2} & a_{n-4} \\
        s^{n-1} & a_{n-1} & a_{n-3} & a_{n-5} \\
        s^{n-2} & b_{n-1} & b_{n-3} & b_{n-5} \\
        s^{n-3} & c_{n-1} & c_{n-3} & n_{n-5} \\
        \vdots & \vdots & \vdots & \vdots \\
        s^{0} & h_{n-1} & &
    \end{array}
\end{equation}

Onde,

\begin{equation}
    \begin{array}{l}
        b_{n-1}=\frac{a_{n-1} a_{n-2}-a_{n} a_{n-3}}{a_{n-1}}=\frac{-1}{a_{n-1}}\left|
        \begin{array}{ll}
            a_{n} & a_{n-2} \\
            a_{n-1} & a_{n-3}
        \end{array}\right| \\
            b_{n-3}=-\frac{1}{a_{n-1}}\left|
        \begin{array}{lc}
            a_{n} & a_{n-4} \\
            a_{n-1} & a_{n-5}
        \end{array}\right| \\
            c_{n-1}=\frac{-1}{b_{n-1}}\left|
        \begin{array}{ll}
            a_{n-1} & a_{n-3} \\
            b_{n-1} & b_{n-3}
        \end{array}\right|
    \end{array}
\end{equation}

e assim por diante. O algoritmo para calcular as entradas na matriz pode ser seguido em uma base determinante ou usando a forma da equação para $b_{n-1}$. 
O critério de Routh-Hurwitz afirma que o número de raízes de $q(s)$ com partes positivas reais é igual ao número de mudanças no sinal da primeira coluna do Matriz Routh. Este critério exige que não haja mudanças no sinal na primeira coluna um para um sistema estável. Este requisito é necessário e suficiente. Quatro casos ou configurações distintas da matriz da primeira coluna devem ser considerados, e cada um deve ser tratado separadamente e requer modificações adequadas de o procedimento de cálculo da matriz: (1) Nenhum elemento na primeira coluna é zero; (2) lá é um zero na primeira coluna, mas alguns outros elementos da linha contendo o zero na primeira coluna são diferentes de zero; (3) há um zero na primeira coluna, e a outra os elementos da linha que contém o zero também são zero; e (4) como no terceiro caso, mas com raízes repetidas no eixo $j\omega$.

## 3. A Estabilidade Relativa de Sistemas de Controle com feedback

A verificação da estabilidade usando o critério de Routh-Hurwitz fornece apenas um par- resposta inicial à questão da estabilidade. O critério de Routh-Hurwitz determina o estabilidade absoluta de um sistema, determinando se alguma das raízes do char- a equação acterística encontra-se na metade direita do plano j. No entanto, se o sistema satisfizer cumpre o critério de Routh-Hurwitz e é absolutamente estável, é desejável determinar a estabilidade relativa; ou seja, é necessário investigar o parente amortecimento de cada raiz da equação característica. A estabilidade relativa de um sistema tem pode ser definido como a propriedade que é medida pela parte real relativa de cada raiz ou par de raízes. Assim, a raiz r2 é relativamente mais estável do que as raízes rh rl5 conforme mostrado na Figura 6.6. A estabilidade relativa de um sistema também pode ser definida em termos dos coeficientes de amortecimento relativos £ de cada par de raízes complexas e, portanto, portanto, em termos de velocidade de resposta e ultrapassagem em vez de tempo de acomodação.

Portanto, a investigação da estabilidade relativa de cada raiz é claramente necessária porque, como visto anteriormente, a localização dos pólos de malha fechada no plano-s determina o desempenho do sistema. Portanto, é imperativo que nós

Porque a estabilidade relativa de um sistema é ditada pela localização das raízes da equação característica, uma primeira abordagem usando uma formulação de plano s é estender o critério de Routh-Hurwitz para verificar a estabilidade relativa. Isso pode ser simplesmente realizado utilizando uma mudança de variável, que desloca o eixo do plano S em ordem para utilizar o critério de Routh-Hurwitz. **Exemplo**

_deslocamento de eixos_

Considere a equação característica de terceira ordem simples

\begin{equation}
    q(s)=s^{3}+4 s^{2}+6 s+4
\end{equation}

Como uma primeira tentativa, deixe $s_n = s + 1/2$ e observe que obtemos um array Routh sem um zero ocorrendo na primeira coluna. No entanto, ao definir a variável deslocada $s_n$ igual a $s + 1$, obtemos

\begin{equation}
    \left(s_{n}-1\right)^{3}+4\left(s_{n}-1\right)^{2}+6\left(s_{n}-1\right)+4=s_{n}^{3}+s_{n}^{2}+s_{n}+1
\end{equation}

Em seguida, a matriz Routh é estabelecida como

\begin{equation}
    \begin{array}{l|ll}
        s_{n}^{3} & 1 & 1 \\
        s_{n}^{2} & 1 & 1 \\
        s_{n}^{1} & 0 & 0^{*} \\
        s_{n}^{0} & 1 & 0
    \end{array}
\end{equation}

Existem raízes no eixo imaginário deslocado que pode ser obtido a partir do polinômio auxiliar

\begin{equation}
    U\left(s_{n}\right)=s_{n}{ }^{2}+1=\left(s_{n}+j\right)\left(s_{n}-j\right)=(s+1+j)(s+1-j)
\end{equation}

O deslocamento do eixo do plano-S para determinar a estabilidade relativa de um sistema é um abordagem muito útil, particularmente para sistemas de ordem superior com vários pares de raízes conjugadas de complexo de circuito fechado.

## 4. A Estabilidade de Sistemas com Variáveis de Estado

A estabilidade de um sistema modelado por um modelo de gráfico de fluxo variável de estado pode ser prontamente averiguado. A estabilidade de um sistema com uma função de transferência de entrada-saída $T(s)$ pode ser determinado examinando o polinômio denominador de $T(s)$. Portanto, se a função de transferência é escrita como

\begin{equation}
    T(s) = \frac{p(s)}{q(s)},
\end{equation}

onde $p(s)$ e $q(s)$ são polinômios em $s$, então a estabilidade do sistema está representada pelas raízes de $q(s)$. O polinômio $q(s)$, quando definido igual a zero, é chamado de equação característica. As raízes da equação característica devem estar no semiplano s da esquerda para que o sistema apresente uma resposta temporal estável. Portanto, para verificar a estabilidade de um sistema representado por uma função de transferência, investigamos o equação característica e utilizar o critério de Routh-Hurwitz. Se o sistema nós estamos investigando é representado por um modelo de estado de gráfico de fluxo de sinal, obtemos o equação característica avaliando o determinante do gráfico de fluxo. Se o sistema for representado por um modelo de diagrama de blocos, obtemos a equação característica usando os métodos de redução do diagrama de blocos.

**Exemplo**

_Estabilidade de um sistema_

A função de transferência $T(s)$ examinada no exemplo abaixo

\begin{equation}
    T(s)=\frac{2 s^{2}+8 s+6}{s^{3}+8 s^{2}+16 s+6}
\end{equation}

O polinômio característico para este sistema é

\begin{equation}
    q(s)=s^{3}+8 s^{2}+16 s+6
\end{equation}

Este polinômio característico também é facilmente obtido a partir do gráfico de fluxo modelo ou modelo de diagrama de blocos. Usando o critério de Routh-Hurwitz, descobrimos que o sistema é estável e que todas as raízes de $q(s)$ estão no semiplano s da esquerda.

Freqüentemente determinamos o gráfico de fluxo ou modelo de diagrama de blocos diretamente de um conjunto de equações diferenciais de estado. Podemos usar o gráfico de fluxo diretamente para determinar a estabilidade do sistema, obtendo a equação característica do gráfico de fluxo determinante $\Delta(s)$. Da mesma forma, podemos usar a redução do diagrama de blocos para definir os caracteres equação acterística. Uma ilustração dessas abordagens ajudará na compreensão esses métodos.

**Exemplo**

_Estabilidade de um sistema de segunda ordem_

Um sistema de segunda ordem é descrito pelas duas equações diferenciais de primeira ordem

\begin{equation}
    \dot{x}_{1}=-3 x_{1}+x_{2} \text { and } \dot{x}_{2}=+1 x_{2}-K x_{1}+K u
\end{equation}

Usando a fórmula de ganho de fluxo de sinal de Mason, notamos três loops:

\begin{equation}
    L_{1}=s^{-1}, \quad L_{2}=-3 s^{-1}, \text { and } L_{3}=-K s^{-2},
\end{equation}

onde $L_1$ e $L_2$ não compartilham um nó comum. Portanto, o determinante

\begin{equation}
    \Delta=1-\left(L_{1}+L_{2}+L_{3}\right)+L_{1} L_{2}=1-\left(s^{-1}-3 s^{-1}-K s^{-2}\right)+\left(-3 s^{-2}\right) .

Como todos os coeficientes devem ser positivos, exigimos $K > 3$ para estabilidade. Uma análise similar pode ser realizada usando o diagrama de blocos. Fechando os dois ciclos de feedback produz as duas funções de transferência

\begin{equation}
G_{1}(s)=\frac{1}{s-1} \text { e } G_{2}(s)=\frac{1}{s+3}

conforme ilustrado na Figura (b). A função de transferência de malha fechada é, portanto,

\begin{equation}

Portanto, a equação característica é

\begin{equation}
    \Delta(s)=1+K G_{1}(s) G_{2}(s)=0
\end{equation}

ou

\begin{equation}
    \Delta(s)=(s-1)(s+3)+K=s^{2}+2 s+(K-3)=0
\end{equation}

Isso confirma os resultados obtidos usando técnicas de gráfico de fluxo de sinal.

Um método de obtenção da equação característica diretamente do vetor dif- equação ferencial é baseada no fato de que a solução para o sistema não forçado é um função exponencial. A equação diferencial vetorial sem sinais de entrada é

$$ x = Ax,$$

onde $x$ é o vetor de estado. A solução é de forma exponencial, e podemos definir um constante $\lambda$ tal que a solução do sistema para um estado pode ser da forma $X_i(t) = k_ie^{\lambda , t}$. O $\lambda_i$, são chamados de raízes características ou autovalores do sistema, que são simplesmente as raízes da equação característica. Se deixarmos $x = ke^{\lambda t}$ e substituindo na equação acima, temos

$$ \lambda ke^{\lambda t} = Ake^{\lambda t} $$,

ou

$$ \lambda x = Ax $$.

A equação pode ser reescrita como

$$ (\lambda I - A) x = 0 $$

onde I é igual à matriz identidade e 0 é igual à matriz nula. Este conjunto de novas equações têm uma solução não trivial se e somente se o determinante desaparecer - isto é, somente se

$$ det(\lambda I - A) = 0 $$

A equação de ordem énesima em $\lambda$ resultante da avaliação deste determinante é a equação característica e a estabilidade do sistema podem ser facilmente verificadas.

**Exemplo**

_Sistema epidêmico fechado_

A equação diferencial vetorial do sistema epidêmico é dada aqui como

\begin{equation}
    \frac{d \mathbf{x}}{d t}=\left[
    \begin{array}{rrr}
        -\alpha & -\beta & 0 \\
        \beta & -\gamma & 0 \\
        \alpha & \gamma & 0
    \end{array}\right]
        \mathbf{x}+\left[
    \begin{array}{ll}
        1 & 0 \\
        0 & 1 \\
        0 & 0
    \end{array}
        \right]\left[
    \begin{array}{l}
        u_{1} \\
        u_{2}
    \end{array}\right]
\end{equation}

A equação característica é então

\begin{equation}
    \begin{aligned}
            \operatorname{det}(\lambda \mathbf{I}-\mathbf{A}) &=\operatorname{det}\left\{\left[
        \begin{array}{lll}
            \lambda & 0 & 0 \\
            0 & \lambda & 0 \\
            0 & 0 & \lambda
        \end{array}
            \right]-\left[
        \begin{array}{rrr}
            -\alpha & -\beta & 0 \\
            \beta & -\gamma & 0 \\
            \alpha & \gamma & 0
        \end{array}
            \right]\right\} \\&=\operatorname{det}\left[
        \begin{array}{ccc}
            (\lambda+\alpha) & \beta & 0 \\
            -\beta & (\lambda+\gamma) & 0 \\
            -\alpha & -\gamma & \lambda
        \end{array}\right] \\
            &=\lambda\left[(\lambda+\alpha)(\lambda+\gamma)+\beta^{2}\right] \\
            &=\lambda\left[\lambda^{2}+(\alpha+\gamma) \lambda+\left(\alpha \gamma+\beta^{2}\right)\right]=0
    \end{aligned}
\end{equation}

Assim, obtemos a equação característica do sistema, e é semelhante àquela obtido por métodos de fluxograma. A raiz adicional $\lambda = 0$ resulta da definição de $x3$ como a integral de $\alpha x_1 + \gamma x_2$, e $x_3$ não afeta as outras variáveis de estado. Assim, a raiz A = 0 indica a integração conectada com $x3$. A equação característica indica que o sistema é marginalmente estável quando $(\alpha + \gamma) > 0$ e  $(\alpha \gamma + \beta^2) > 0$.

Como outro exemplo, considerando o exemplo de um pêndulo invertido. A matriz do sistema é

\begin{equation}
    \mathbf{A}=\left[
    \begin{array}{cccc}
        0 & 1 & 0 & 0 \\
        0 & 0 & -m g / M & 0 \\
        0 & 0 & 0 & 1 \\
        0 & 0 & g / l & 0
    \end{array}\right]
\end{equation}

A equação característica pode ser obtida a partir do determinante de $(\lambda I - A)$ como segue:

\begin{equation}
    \operatorname{det}\left[
    \begin{array}{cccr}
        \lambda & -1 & 0 & 0 \\
        0 & \lambda & m g / M & 0 \\
        0 & 0 & \lambda & -1 \\
        0 & 0 & -g / l & \lambda
    \end{array}
    \right]=\lambda\left[\lambda\left(\lambda^{2}-\frac{g}{l}\right)\right]=\lambda^{2}\left(\lambda^{2}-\frac{g}{l}\right)=0 .
\end{equation}

A equação característica indica que existem duas raízes em $\lambda = 0:$ uma raiz em $\lambda = +\sqrt{g/l}$ e uma raiz em $\lambda = -\sqrt{g/l}$. Portanto, o sistema é instável, porque há uma raiz no plano da direita em $\lambda = +\sqrt{g/l}$. As duas raízes em $\lambda = 0$ também resultarão em uma resposta ilimitada.
